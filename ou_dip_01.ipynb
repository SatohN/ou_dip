```
# 画像の読み込み
img = cv2.imread("sample.jpg",cv2.IMREAD_GRAYSCALE)
```

```
# Caution: You should NOT use many FORs in Python scripts!!
# 以下は悪い例：Python 人生は、いかにfor文を書かないかに懸かっている。
out = img.copy()
for y in range(out.shape[0]):
  for x in range(out.shape[1]):
    if out[y][x] < 128:
      out[y][x] = 255
```

```
# 上記は、numpyを使って以下のようにも実現できる
out = np.where(img<128, 255, img)
```

```
cv2_imshow(out) # 画像の表示
cv2.imwrite("output.jpg",out) # 画像の保存
```
